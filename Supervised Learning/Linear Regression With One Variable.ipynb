```python
# Function to compute the cost
def compute_cost(x, y, w, b):
    m = len(y)
    predictions = w * x + b
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    return cost
```

```python
def gradient_descent(X, y, w, b, alpha, num_iterations):
    m = len(y)
    cost_history = []

    for _ in range(num_iterations):
        predictions = X.dot(w) + b
        errors = predictions - y

        w -= alpha * (1/m) * X.T.dot(errors)  # Update weights w
        b -= alpha * (1/m) * np.sum(errors)  # Update bias b

        cost_history.append(compute_cost(X, y, w, b))

    return w, b, cost_history
```

```python
# Initial parameters (w = 0, b = 0)
w = 0  # Weight (slope)
b = 0  # Bias (intercept)
alpha = 0.1  # Learning rate
num_iterations = 1000  # Number of iterations for gradient descent

# Perform gradient descent to find optimal w and b
w_optimal, b_optimal, cost_history = gradient_descent(x, y, w, b, alpha, num_iterations)
```

```python
# Plotting the cost function over iterations
plt.plot(cost_history)
plt.title("Cost Function over Iterations")
plt.xlabel("Iteration")
plt.ylabel("Cost (J)")
plt.show()
```

```python
print(f"Optimal Parameters: w = {w_optimal}, b = {b_optimal}")

# Let's plot the data and the best-fit line
plt.scatter(x, y, color='red', label='Training Data')
plt.plot(x, w_optimal * x + b_optimal, label='Linear Regression', color='blue')
plt.title("Size vs Price")
plt.xlabel("Size of House (1000 sqft)")
plt.ylabel("Price of House ($100,000)")
plt.legend()
plt.show()
```
